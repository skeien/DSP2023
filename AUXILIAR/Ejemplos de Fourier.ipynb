# Ejemplos de Fourier:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,1,100)
f = 5
phi = 0.5*np.pi

seno = np.sin(2*np.pi*f*t+phi)
plt.title('Sinusoidal 5 Hz')
plt.xlabel('Tiempo [s]')
plt.ylabel('Magnitud')
plt.plot(t,seno)
plt.show()
```

```python
F_seno = np.fft.fft(seno)
freq = np.linspace(0,100,100)
plt.plot(freq,abs(F_seno))
plt.title('TDF sinusoidal 5 Hz')
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud')
plt.grid()
```

```python
impulse = np.zeros(100)
impulse[0] = 1 
f_impulse = np.fft.fft(impulse)
plt.plot(freq,abs(f_impulse))

plt.title('TDF Impulso')
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud')
plt.grid()
plt.show()
```

```python

impulse = np.zeros(100)
impulse[10] = 1 
f_impulse = np.fft.fft(impulse)
x = freq
y = f_impulse.real
z = f_impulse.imag

plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title('TDF Impulso')
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud - Real')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(x,z)
plt.xlabel('Frecuencia')
plt.ylabel('Magnitud - Fase')
plt.grid()
plt.show()

```

```python
%matplotlib notebook
y0= np.zeros(100)
z0= np.zeros(100)
ax = plt.axes(projection='3d')
ax.plot3D(x, y0, z0, 'red')
ax.plot3D(x, y, z, 'blue')
ax.set_title('Impulso con fase desplazada')
ax.set_xlabel('Frecuencia')
ax.set_ylabel('Magnitud - Fase')
ax.set_zlabel('Magnitud - Real')
plt.show()
```
