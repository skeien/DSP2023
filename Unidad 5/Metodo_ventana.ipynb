```python
n=np.arange(-40,40+1,1)
```

```python
h=(wc/np.pi)*np.sinc(wc*n/np.pi)
plt.stem(h)

```

```python
window=np.hamming(81)
plt.stem(window)
```

```python
filt=h*window
plt.stem(filt)
```

```python
X=np.fft.fft(filt,1000)
fs=1000
freqs=np.linspace(0,fs,len(X))
plt.plot(freqs,abs(X))
```

```python
fs=1000
f1=320
f2=110
t=np.arange(0,0.2,1/fs)
y=np.sin(2*np.pi*f1*t)+np.sin(2*np.pi*f2*t)

plt.plot(y)

Y=np.fft.fft(y)
freqs=np.linspace(0,fs,len(y))
```

```python
plt.plot(freqs,abs(Y))
```

```python
yf=np.convolve(y,filt)

plt.plot(yf)

```

```python
YF=np.fft.fft(yf)
freqs=np.linspace(0,fs,len(yf))


plt.plot(freqs,abs(YF))
```
